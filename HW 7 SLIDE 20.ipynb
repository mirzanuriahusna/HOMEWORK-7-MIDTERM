```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
#data 1
#Mengambil data CSV
data = np.genfromtxt('1.csv', delimiter=',', usecols=(1, 2, 3, 4), dtype='int32')

#Mengambil subset data (1000-1100)
y = data[1000:1100, 1]

#Menemukan puncak dengan jarak tertentu
peaks_d10, _ = find_peaks(y, distance=10)
peaks_d3, _ = find_peaks(y, distance=3)

#Mencetak indeks puncak yang ditemukan
print('distance 10 idx:', peaks_d10)
print('distance 3 idx:', peaks_d3)

#Menampilkan grafik
plt.figure(figsize=(12, 5))
plt.plot(y, label='Data')
plt.plot(peaks_d3, y[peaks_d3], "vr", label='distance 3')
plt.plot(peaks_d10, y[peaks_d10], "xk", label='distance 10')
plt.legend()
plt.title('Puncak dengan Distance 3 dan 10')
plt.xlabel('Indeks')
plt.ylabel('Nilai')
plt.grid(True)
plt.show()  
```
